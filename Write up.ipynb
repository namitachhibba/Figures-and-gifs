<b> COVID-19 Study: A Data Analytics Project </b>

### Introduction

Understanding of COVID-19 data educates the people, organizations and the governments of the countries about the development  and progression of the disease in the world and in their country. It is very important to be aware about how steadily and severely the virus is advancing, to cope with it by updating the stringency of the regulatory laws on businesses, people’s movements and travel. 

This project aims to understand the COVID-19 severity worldwide by analyzing the country wise data for COVID-19 in terms of cumulative confirmed cases, deaths, recovered cases and testing count. The information regarding population and population density for the countries is also included to understand their correlation if any with the spread of disease and its impact.

As in any data analytics project, this project involves three phases. 
- Phase 1: Data Preparation
- Phase 2: Data Exploration
- Phase 3: Data Modelling

So let’s begin our step by step journey to uncover COVID-19 situations in individual countries and their comparison with each other.



### Data Preparation <i> (Phase 1) </i>

Data preparation is the first step in data analytics projects and can include many discrete tasks such as loading data, data fusion, data cleaning, data augmentation, data transformation and data delivery.

Our project will involve the majority of these tasks viz.
- <b> Step 1: Loading Data: </b>

<i> Copying and loading data sets from source files </i>


- <b> Step 2: Data Cleaning </b>

<i> Detecting and correcting corrupt or inaccurate records from a table, or dataframe and identifying incomplete, incorrect, inaccurate or irrelevant parts of the data and then replacing, modifying, or deleting the dirty or coarse data. </i>

- <b> Step 3: Data Fusion:  </b>

<i> Integrating multiple data sources to produce more consistent, accurate, and useful information than that provided by any individual data source.</i>

- <b> Step 4: Data Transformation </b>

<i> Converting data from one format or structure into another format or structure. </i>

#### Step 4: Data Transformation

##### Deriving Daily Confirmed, Deaths, Recovered, Testing counts from Cumulative Confirmed, Deaths, Recovered, Testing counts using existing columns in MergedData dataframe 

Doing differencig we can deduce the daily counts from cumulative counts. For example for "Daily Confirmed Count" we apply differencing on "Cumulative Confirmed Count" as follows:

> <font color = blue> MergedData['Daily Confirmed Count'] = MergedData['Cumulative Confirmed Count'].diff().fillna(MergedData["Cumulative Confirmed Count"]) </font> <br> <br> 

And we do the same for "Daily Death Count", "Daily Recovered Count" and "Daily Testing Count"

<br> <br>

#####  Deriving Active Cases, Recovery Rate and Testing Ratios (Cumulative Testing per Population, Daily Testing per 1000, Daily Testing Count per Daily Confirmed Count) from the available columns in the MergedData using following formula

- <font color = blue> "ActiveCases" = "Cumulative Confirmed Count" - "Cumulative Recovered Count" - "Cumulative Death Count" </font> <br>

- <font color = blue> "RecoveryRate" = "Cumulative Recovered Count" / "Cumulative Confirmed Count </font> <br>

- <font color = blue> "CumTesting_per_Population"= "Cumulative Testing Count" / "Population"</font> <br>

- <font color = blue> "DailyTesting_per_1000" = "Daily Testing Count"/ 1000 </font> <br>
